#### 1. Plotting a Simple Line Chart

```julia
using Plots

# Data
x = 1:10
y = x .^ 2

# Plot
plot(x, y, label="y = x^2", xlabel="X-axis", ylabel="Y-axis", title="Simple Line Plot")
```

#### 2. Plotting Multiple Series in One Graph


```julia
# Additional data
y2 = x .^ 3

# Plot multiple series
plot(x, y, label="y = x^2", xlabel="X-axis", ylabel="Y-axis", title="Multiple Series Plot")
plot!(x, y2, label="y = x^3")  # Use `plot!` to add the second series
```

#### 3. Creating a Scatter Plot


```julia
# Random data
x = rand(10)
y = rand(10)

# Scatter plot
scatter(x, y, label="Points", xlabel="Random X", ylabel="Random Y", title="Scatter Plot")
```

#### 4. Creating a Bar Plot


```julia
# Data
categories = ["A", "B", "C", "D"]
values = [5, 3, 8, 6]

# Bar plot
bar(categories, values, label="Values", title="Bar Plot", xlabel="Categories", ylabel="Values")
```

#### 5. Creating a Histogram


```julia
# Data
data = randn(1000)  # Random data with a normal distribution

# Histogram
histogram(data, bins=30, title="Histogram", xlabel="Values", ylabel="Frequency")
```

#### 6. Creating a Pie Chart

```julia
# Data
values = [40, 25, 20, 15]
labels = ["A", "B", "C", "D"]

# Pie chart
pie(labels, values, title="Pie Chart")
```

#### 7. Adding Custom Styles (Color, Thickness, Line Type)
##### lw: Line width
##### lc: Line color
##### ls: Line style (:dash, :dot, :solid, etc.)

```julia
x = 0:0.1:10
y = sin.(x)

# Custom plot
plot(x, y, label="sin(x)", lw=3, lc=:blue, ls=:dash, xlabel="X", ylabel="sin(x)", title="Customized Plot")
```

#### 8. Creating a 3D Plot

```julia
# Data
x = 1:10
y = 1:10
z = x .* y'

# 3D surface plot
surface(x, y, z, title="3D Surface Plot", xlabel="X", ylabel="Y", zlabel="Z")
```

#### 9. Creating a Heatmap

```julia
# Data
matrix = rand(10, 10)

# Heatmap
heatmap(matrix, title="Heatmap", xlabel="Columns", ylabel="Rows")
```
