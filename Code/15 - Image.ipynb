```julia
using Images  # Import the Images package to handle image operations
using ImageFiltering
# Load the image from the file path
img = Images.load("D:/Learn/Julia Programming/Attachment/File Handling/sample_640×426.jpg")

# Display the image
display(img)
```

```julia
dimensions = size(img)
println("Dimensions: ", dimensions)
```

```julia
data_type = typeof(img)
println("Data Type: ", data_type)
```

```julia
# Adjust the crop range to fit within the image dimensions
cropped_img = img[50:150, 50:200, :]

# Display the cropped image
display(cropped_img)
```

```julia
# Save the cropped image to a PNG file
save("D:/Learn/Julia Programming/Attachment/File Handling/cropped_image.png", cropped_img)
```
